# 03.Sample Exercise Steps

```python
#creating user defined function to sort products into price ranges
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#applying the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
df['prices'].max()
```

```python
#run loc function to assign values based on price
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#count new labels, same as user defined function
df['price_range_loc'].value_counts(dropna=False)
```

### Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
#creating new variable for busiest days
result_new=[]

for value in df_ords_prods["order_day_of_week"]:
  if value in [0,1]:
    result_new.append("Busiest days")
  elif value in [4,3]:
    result_new.append("Slowest days")
  else:
    result_new.append("Regularly busy")
```

```python
#creating new column to put the list
df_ords_prods['busiest_days']=result_new
```

### When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
#Checking which hour of day most orders take place
df_ords_prods['order_hour_of_day'].value_counts(dropna=False)
```

```python
#creating variable for busiest times of day
result_hour=[]

for value in df_ords_prods["order_hour_of_day"]:
  if value>=9 and value<=16:
    result_hour.append("Most orders")
  elif value<=6:
    result_hour.append("Fewest orders")
  else:
    result_hour.append("Average orders")
```

```python
#creating new column to put the list
df_ords_prods['busiest_period_of_day']=result_hour
```

### Print the frequency for this new column.

```python
#Checking count of results in new column
df_ords_prods['busiest_period_of_day'].value_counts(dropna=False)
```
