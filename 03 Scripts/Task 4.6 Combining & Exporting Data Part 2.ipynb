# 03.Combining Dataframes

### Determine a suitable way to combine the orders_products_combined dataframe with your products data set. Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task.

```python
#Importing products_cleaned.csv
df_prod=pd.read_csv(os.path.join(path,'02 Data', 'Prepared Data', 'products_cleaned.csv'))
```

```python
df_prod.shape
```

```python
df_ord_prod.head()
```

```python
df_prod = df_prod.drop(columns = ['Unnamed: 0'])
```

```python
df_prod.head()
```

#### Only common column is product_id so that is how they will need to be combined.

```python
df_ord_prod_merged=df_ord_prod.merge(df_prod, on = 'product_id')
```

### Confirm the results of the merge using the merge flag.
