# 03.Data Consistency Checks

```python
df_prods
```

```python
#Checking for missing values
df_prods.isnull().sum()
```

```python
# checking for mixed-type columns
for col in df_prods.columns.tolist():
  weird = (df_prods[[col]].applymap(type) != df_prods[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_prods[weird]) > 0:
    print (col)
```

```python
#creating dataframe of only missing value from products.csv
df_prods_nan=df_prods[df_prods['product_name'].isnull()==True]
```

```python
df_prods_nan
```

```python
df_prods.shape
```

```python
#Cleaning missing value rows out of df_prods 
df_prods_clean = df_prods[df_prods['product_name'].isnull() == False]
```

```python
#checking that 16 rows are now removed
df_prods_clean.shape
```

```python
#Checking for duplicates in df_prods_clean
df_prods_dups=df_prods_clean[df_prods_clean.duplicated()]
```

```python
df_prods_dups
```

```python
#removing duplicates
df_prods_clean_no_dups=df_prods_clean.drop_duplicates()
```

```python
df_prods_clean_no_dups.shape
```

```python
# Exporting df_prods_clean_no_dups dataframe as “products_cleaned.csv”
df_prods_clean_no_dups.to_csv(os.path.join(path,'02 Data','Prepared Data','products_cleaned.csv'))
```
