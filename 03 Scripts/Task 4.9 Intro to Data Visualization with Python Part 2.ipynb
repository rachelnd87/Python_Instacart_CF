```python
#Creating a list holding true/false values to make a 70/30 split on data
np.random.seed(4)
dev=np.random.rand(len(df)) <= 0.7
```

```python
#Store 30% of the data in the dataframe small
small = df[~dev]
```

```python
#Checking both samples equal total
len(df)
```
