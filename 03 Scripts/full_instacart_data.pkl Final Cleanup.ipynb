# 01.Importing Libraries

```python
import pandas as pd
import numpy as np
import os
```

# 02.Importing Data

```python
#defining path
path=r'C:\Users\rache\Documents\Instacart Basket Analysis'
```

```python
#importing full_instacart_data.pkl
df=pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'full_instacart_data.pkl'))
```

# 03.Dropping Columns

```python
df= df.drop(columns = ['_merge'])
```

```python
df= df.drop(columns = ['Unnamed: 0'])
```

```python
df.dtypes
```

# 04.Renaming Columns

```python
#renaming columns
df.rename(columns = {'First Name' : 'first_name'}, inplace = True)
df.rename(columns = {'Last Name' : 'last_name'}, inplace = True)
df.rename(columns = {'Gender' : 'gender'}, inplace = True)
df.rename(columns = {'State' : 'state'}, inplace = True)
df.rename(columns = {'Age' : 'age'}, inplace = True)
df.rename(columns = {'No. of Dependents' : 'no_of_dependents'}, inplace = True)
```

```python
df.dtypes
```

# 05.Changing Data Types

```python
#changing data type for columns to reduce memory usage
df['order_id'] = df['order_id'].astype('str')
df['product_id'] = df['product_id'].astype('str')
```

```python
df['order_number'] = df['order_number'].astype('int8')
df['order_day_of_week'] = df['order_day_of_week'].astype('int8')
df['order_hour_of_day'] = df['order_hour_of_day'].astype('int8')
df['product_id'] = df['product_id'].astype('int32')
df['add_to_cart_order'] = df['add_to_cart_order'].astype('int8')
df['reordered'] = df['reordered'].astype('int8')
df['aisle_id'] = df['aisle_id'].astype('int16')
df['department_id'] = df['department_id'].astype('int8')
df['prices'] = df['prices'].astype('float16')
df['days_since_last_order'] = df['days_since_last_order'].astype('float16')
df['avg_price'] = df['avg_price'].astype('float16')
df['median_order_frequency'] = df['median_order_frequency'].astype('float16')
```

```python
df.dtypes
```

```python
df['product_id'] = df['product_id'].astype('str')
```

```python
df.dtypes
```

```python
#Exporting df dataframe as “full_instacart_data_final.pkl”
df.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'full_instacart_data_final.pkl'))
```
