### Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
#create a new column for maximum orders for each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
#deriving columns with loc function
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
#checking flags assigned correctly
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

### The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). 

```python
#grouping by loyalty flag and creating basic statistics of prices of products purchased by each type of customer
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### Create a spending flag for each user based on the average price across all their orders using the following criteria: If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.” & If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#create a new column for average prices for each user
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#deriving columns with loc function
ords_prods_merge.loc[ords_prods_merge['avg_price'] <10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] >=10, 'spending_flag'] = 'High spender'
```

```python
#checking frequency
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
#checking flags assigned correctly
ords_prods_merge[['user_id', 'spending_flag', 'avg_price']].head(60)
```

###  Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows: If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”, If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”, & If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#create a new column for order frequency for each user
ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
#deriving columns with loc function
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] >20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] >10) & (ords_prods_merge['median_order_frequency'] <=20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <=10, 'frequency_flag'] = 'Frequent customer'
```

```python
#checking frequency
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
#checking flags assigned correctly
ords_prods_merge[['user_id', 'frequency_flag', 'median_order_frequency']].head(60)
```
